```python
# импорт библиотек и начальные настройки
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations
from scipy.stats import ttest_ind

pd.set_option('display.max_rows', 50)  # показывать больше строк
pd.set_option('display.max_columns', 50)  # показывать больше колонок
```

```python
# процедуры и функции
# процедура первичного анализа данных в зависимости от типа
def dfanalize(s):
    print('Переменная', s.name)
    print('Число уникальных значений:', s.nunique())

    # смотрим начальное распределение значений
    print('Начальное распределение значений переменной:')
    display(s.describe())

    # определение границ выбровсов для числовых переменных:
    if s.dtype in ['float64', 'int64']:
        print('Значения находятся в диапазоне от', s.min(), 'до', s.max())
        perc25 = s.quantile(0.25)
        perc75 = s.quantile(0.75)
        IQR = perc75 - perc25
        print('Большая часть значений лежит между', perc25, 'и', perc75,
              '(25-й перцентиль: {},'.format(perc25), '75-й перцентиль: {})'.format(perc75))
        print('IQR: {}, '.format(IQR), 'Границы выбросов: [{f}, {l}].'.format(
            f=perc25 - 1.5*IQR, l=perc75 + 1.5*IQR))
        # выведем гистограммы:
        s.hist(bins=s.nunique(), label=s.name)
        s.loc[s.between(perc25 - 1.5*IQR, perc75 + 1.5*IQR)
              ].hist(bins=s.nunique(), label='IQR')

    # обработка номинативных переменных
    if s.dtype == 'object':
        s.hist(label=s.name)
        print('Наиболее частое значение:', s.value_counts().idxmax())
        display(pd.DataFrame(s.value_counts()))
        print("Значений, встретившихся в столбце более 100 раз:",
              (s.value_counts() > 100).sum())
    plt.legend()

# процедура для построения box-plot
def get_boxplot(column):
    fig, ax = plt.subplots(figsize=(4, 3))
    sns.boxplot(x=column,
                y='score',
                data=sm.loc[sm.loc[:, column].isin(
                    sm.loc[:, column].value_counts().index[:])],
                ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()

# процедура для теста Стьдента
def get_stat_dif(column):
    cols = sm.loc[:, column].value_counts().index[:]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(sm.loc[sm.loc[:, column] == comb[0], 'score'],
                     sm.loc[sm.loc[:, column] == comb[1], 'score']).pvalue \
                <= 0.05/len(combinations_all):  # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
```

```python
# прочитаем файл и выведем основную информацию о данных
sm = pd.read_csv('stud_math.csv')
display(sm.head(10))
sm.info()
```

```python
# преобразование данных
# приведем название столбцов к единому стилю:
sm.rename(columns=lambda x: x.lower(), inplace=True)

# переменная 'studytime, granular' представляет собой столбец  studytime, умноженный на коэффициент -3 и поэтому он не нужен для анализа
sm['studytime, granular'].value_counts()
sm['studytime'].value_counts()
print('correlation coef: ', sm['studytime, granular'].corr(sm['studytime']))

# удалим переменную 'studytime, granular'
sm.drop(['studytime, granular'], inplace=True, axis=1)

# преобразуем номенативные переменные с бинарными значениями yes/no к числовому типу: yes=1, no =0
for i in ['schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']:
    sm[i] = sm[i].apply(lambda x: None if pd.isnull(x)
                        else 1 if x == 'yes' else 0)
sm.info()
# число количественных признаков  - 19+1 =20
# число номинативных переменных - 9
# рассмотрим числовые переменные:
```

Рассмотрим данные переменные более детально:

```python
# обработаем fedu
display(sm.fedu.value_counts())
# есть ошибочное значение: 40, возможно значение данного наблюдения должно было равняться 4.
# т.к. значение 4 встречается довольно часто, можем заменить ошибку на значение 4
sm.fedu = sm['fedu'].apply(lambda x: 4 if x == 40.0 else x)

# теперь ошибка удалена
display(sm.fedu.describe())
sm.fedu.hist(bins=sm.fedu.nunique(), label='fedu')
plt.legend()
```

```python
# обработаем  famrel
display(sm.famrel.value_counts())

# есть ошибочное значение: -1, возможно значение данного наблюдения должно было равняться 1.
# т.к. значение 1 редкое, заменять на него ошибку не будем
# отфильтруем данные исходя из здравого смысла, строки с пустыми значениями оставляем:
sm = sm.loc[(sm.famrel.between(1, 5) | pd.isnull(sm.famrel))]

# теперь ошибка удалена
display(sm.famrel.describe())
sm.famrel.hist(bins=sm.famrel.nunique(), label='famrel')
plt.legend()
```
